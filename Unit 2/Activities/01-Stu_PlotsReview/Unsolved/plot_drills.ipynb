```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA SET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
xax = np.arange(len(gyms))
```

```python
plt.bar(xax,members)
```

```python
# DATA SET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)

plt.plot(x_axis,sin)
plt.show()
```

```python
# DATA SET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.pie(members,explode=explode,labels=gyms,colors=colors,textprops={'color':"w", 'size':20})
plt.figure()
plt.show()
```

```python
# DATA SET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
plt.plot(x_axis,times)
```
